```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Get the file path for running team one
team_one_file_path = os.path.join('../','Resources', 'team-1.csv')
#team_two_file_path = os.path.join('../','Resources', 'team-2.csv')
```

```python
# Read the CSV file into a DataFrame
team_one_df = pd.read_csv(team_one_file_path)
#team_two_file_df = pd.read_csv(team_two_file_path)
```

```python
# Melt the DataFrame
team_one_df.melt()
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
team_one_df1 = team_one_df.melt(id_vars = "runner_name")
team_one_df1.head()
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
team_one_df.melt(id_vars = "coach_name")
```

```python
# Renaming the columns
team_one_df1 = team_one_df1.rename(columns={"runner_name":"runner_name", "variable":"variable", "value":"length"})
team_one_df1.head()
```

```python
# Remove coach name from DataFrame
index_names = team_one_df1[(team_one_df1["variable"]=="coach_name")].index
team_one_df1.drop(index_names, inplace = True)
team_one_df1.head()
```

```python
# Show final result
results_df = team_one_df1.melt(id_vars="runner_name", var_name="run",value_name="length")
```

```python
# Save results to a file
results_df.to_csv('melted_team_one_df1.txt', sep = ',')
```
