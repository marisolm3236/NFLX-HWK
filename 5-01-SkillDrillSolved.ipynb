```python
# Dependencies
import pandas as pd
import plotly.express as plt
import os
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
```

```python
bikecounts_file = os.path.join('../','Resources', 'bikecounts.csv')
bikecounts_df = pd.read_csv(bikecounts_file)
bikecounts_df['Day'] = pd.to_datetime(bikecounts_df['Day'])
bikecounts_df.head()
```

```python
def remove_unwanted(x):
    for unwanted_char in ['T', '(S)', ' ']:
        x=x.replace(unwanted_char, '')
    return x
bikecounts_df['Precipitation']=bikecounts_df['Precipitation'].apply(remove_unwanted)
bikecounts_df['Precipitation']=bikecounts_df['Precipitation'].replace('',0).astype('float64')
bikecounts_df.head()
```

```python
# Select your independent X terms, and your dependent y term. 
X = bikecounts_df[['Precipitation','High Temp (°F)']]
y = bikecounts_df['Brooklyn Bridge']
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
```

```python
# Print the linear regression results
print(results.summary())
```

```python
coef = results.params
print(f"The equation of our line is y = (Precipitation*{coef[1]}) + (High Temp (°F) *{coef[2]})+ {coef[0]}")
```

* Hypothesis - If we assume that Precipitation and High Temp (°F) has an effect on the number of bikes crossing Brooklyn Bridge then less bikes will cross Brooklyn Bridge.

* Null Hypothesis - If we assume that Precipitation and High Temp (°F) has an effect on the number of bikes crossing Brooklyn Bridge then there is not change.

* R-Squared value of 0.681 would indicate that about 70% of the variance of the dependent variable being studied is explained by the variance of the independent variable
